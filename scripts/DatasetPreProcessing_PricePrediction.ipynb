<h5>Downloading the NFT's and storing them in the appropriate format</h5>

```python
def download_and_save_image(image_url, image_filename):
    response = requests.get(image_url)
    if response.status_code == 200:
        with open(image_filename, 'wb') as f:
            f.write(response.content)
        return True
    else:
        return False
```

<h5>Filtering out the NFT's that have an incorrect format </h5>

```python
import json
import os


# Create a dictionary to store image file path - price_in_eth mapping
data = {}
for image_path in X:
    image_id = os.path.splitext(os.path.basename(image_path))[0]
    row = assets_df[assets_df['id'] == int(image_id)]
    if not row.empty:
        price_in_eth = row.iloc[0]['total_price_in_eth']
        data[image_path] = price_in_eth

# Save the dictionary as a JSON file
json_file = '/content/drive/MyDrive/NFT_Images/image_price_mapping.json'
with open(json_file, 'w') as f:
    json.dump(data, f)


print("Image-price mapping saved as JSON:", json_file)
```

```python
import cv2
import numpy as np
import imghdr

log_file_path = "/content/drive/MyDrive/Capstone-DatasetPrep/logs.txt"

def load_image(file_path):
    try:
        # Detect the image format using imghdr
        image_format = imghdr.what(file_path)
        if image_format in ['jpeg', 'jpg']:
            image = cv2.imread(file_path)
            return image
        elif image_format == 'png':
            image = cv2.imread(file_path)
            return image
        elif image_format == 'gif':
            from PIL import Image
            image = Image.open(file_path)
            return np.array(image)
        else:
            with open(log_file_path, "a") as log_file:
                log_file.write(f"Unsupported file format for {file_path}: {image_format}\n")
            return None
    except Exception as e:
        with open(log_file_path, "a") as log_file:
            log_file.write(f"Error loading image from {file_path}: {e}\n")
        return None

image_paths = X

images = []
for file_path in image_paths:
    image = load_image(file_path)
    if image is not None:
        images.append(image)

print(images)
```

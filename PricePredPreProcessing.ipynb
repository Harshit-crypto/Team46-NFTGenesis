<h5>Downloading the NFT's and storing them in the appropriate format</h5>

```python
def download_and_save_image(image_url, image_filename):
    response = requests.get(image_url)
    if response.status_code == 200:
        with open(image_filename, 'wb') as f:
            f.write(response.content)
        return True
    else:
        return False
```

```python
image_dir = '/content/drive/MyDrive/NFT_Images'

for index, row in assets_df.iterrows():
    image_url = row['image_url']
    image_id = row['id']
    total_price_in_eth = row['total_price_in_eth']

    if pd.notna(image_url):  # Check if the 'image_url' is not NaN
          image_filename = os.path.join(image_dir, f"{image_id}.jpg")

          if not os.path.exists(image_filename):
            # Download and save the image
            success = download_and_save_image(image_url, image_filename)

            if success:
                print(f"Image {image_id} saved as {image_filename}")
            else:
                print(f"Failed to download image {image_id}")
          else:
            print(f"Image {image_id} already exists, skipping...")
    else:
        print(f"Image URL missing for {image_id}, skipping...")
```

```python
# Function to load the images
def load_image(image_path):
    img = cv2.imread(image_path)
    return cv2.resize(img, (224, 224))

image_dir = '/content/drive/MyDrive/NFT_Images'

# Extract prices and IDs from the DataFrame
Y = assets_df['total_price_in_eth'].values
ids = assets_df['id'].values

# Retrieve all image file paths and store them in a list
X = []

images = glob.glob('/content/drive/MyDrive/NFT_Images/*.jpg')
for image in images:
    X.append(image)
```

<h5>Creating a JSON for storing the extracted data to Use for the model </h5>

```python
import json
import os


# Create a dictionary to store image file path - price_in_eth mapping
data = {}
for image_path in X:
    image_id = os.path.splitext(os.path.basename(image_path))[0]
    row = assets_df[assets_df['id'] == int(image_id)]
    if not row.empty:
        price_in_eth = row.iloc[0]['total_price_in_eth']
        data[image_path] = price_in_eth

# Save the dictionary as a JSON file
json_file = '/content/drive/MyDrive/NFT_Images/image_price_mapping.json'
with open(json_file, 'w') as f:
    json.dump(data, f)


print("Image-price mapping saved as JSON:", json_file)
```

<h5>Filtering out the NFT's that have an incorrect format </h5>

```python
import cv2
import numpy as np
import imghdr

def load_image(file_path):
    try:
        # Detect the image format using imghdr
        image_format = imghdr.what(file_path)
        if image_format in ['jpeg', 'jpg']:
            # Load JPEG image
            image = cv2.imread(file_path)
            return image
        elif image_format == 'png':
            # Load PNG image
            image = cv2.imread(file_path)
            return image
        elif image_format == 'gif':
            # Load GIF image (you might need additional libraries for this, like PIL)
            # Example using PIL:
            from PIL import Image
            image = Image.open(file_path)
            return np.array(image)
        else:
            print(f"Unsupported file format for {file_path}: {image_format}")
            return None
    except Exception as e:
        print(f"Error loading image from {file_path}: {e}")
        return None

image_paths = X

images = []
for file_path in image_paths:
    image = load_image(file_path)
    if image is not None:
        images.append(image)

print(images)
```
